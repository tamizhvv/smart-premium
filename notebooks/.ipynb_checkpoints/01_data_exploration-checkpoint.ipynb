```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder,OrdinalEncoder
```

```python
train=pd.read_csv("Z:\\smart_premium\\data\\raw\\train.csv")
test=pd.read_csv("Z:\\smart_premium\\data\\raw\\test.csv")
sample_submission=pd.read_csv("Z:\\smart_premium\\data\\raw\\sample_submission.csv")
```

```python
print(f'train_shape:{train.shape}')
print(f'test_shape:{test.shape}')
print(f'submission_shape:{sample_submission.shape}')
```

```python
import matplotlib.pyplot as plt
num_cols_batch1=['Age','Annual Income','Health Score','Premium Amount']

train[num_cols_batch1].hist(bins=100,figsize=(12,8))
plt.tight_layout()
plt.show()
```

```python
for col in num_cols_batch1:
    n_missing=train[col].isnull().sum()
    n_zero=(train[col]==0).sum()
    n_neg=(train[col]<0).sum()
print(f'missing:{n_missing}\n zero:{n_zero} \n n_neg:{n_neg}')
```

```python
for col in ['Age','Annual Income','Health Score']:
    median_val=train[col].median()
    train[col]=train[col].fillna(median_val)
    print(f'filled missing values with median={median_val}')
```

```python
train[num_cols_batch1].isnull().sum()
```

```python
num_cols_batch2=['Number of Dependents','Previous Claims','Vehicle Age','Credit Score','Insurance Duration']
train[num_cols_batch2].hist(bins=50,figsize=(12,8))
plt.tight_layout()
plt.show()
```

```python
for col in num_cols_batch2:
    n_missing=train[col].isnull().sum()
    n_zeros=(train[col]==0).sum()
    n_negs=(train[col] < 0).sum()
    print(f'n_missing---{n_missing},n_zeros---{n_zeros},n_negs----{n_negs}')
```

```python
for col in ['Number of Dependents','Vehicle Age','Credit Score','Insurance Duration']:
    train[col] = pd.to_numeric(train[col], errors='coerce')
```

```python
for col in ['Number of Dependents','Vehicle Age','Credit Score','Insurance Duration']:
    median_val=train[col].median()
    train[col]=train[col].fillna(median_val)
    print(f'the filled missing values are {median_val}')
train['Previous Claims']=train['Previous Claims'].fillna(0)
```

```python
train[num_cols_batch2].isnull().sum()
```

```python
print(train['Credit Score'].min())
print(train['Credit Score'].max())
```

```python
cat_cols_batch1=['Gender','Marital Status','Education Level','Occupation','Location','Policy Type']
for col in cat_cols_batch1:
    print(f'columns--{col}')
    print(f'value counts {train[col].value_counts()}')
    print(f'missing values {train[col].isnull().sum()}')
    print('---------------------')
print(train[cat_cols_batch1].isnull().sum())
```

```python
mode_val=train['Marital Status'].mode()[0]
train['Marital Status']=train['Marital Status'].fillna(mode_val)
train['Occupation']=train['Occupation'].fillna('Unknown')
```

```python
train[cat_cols_batch1].isnull().sum()
```

```python
le=LabelEncoder()
train['Gender']=le.fit_transform(train['Gender'])
```

```python
train['Gender']
```

```python
train['Education Level']
```

```python
edu_order=["High School","Bachelor's","Master's","PhD"]
ord_enc=OrdinalEncoder(categories=[edu_order])
train['Education Level']=ord_enc.fit_transform(train[['Education Level']])
```

```python
train['Education Level']
```

```python
one_hot_cols=['Marital Status','Occupation','Location','Policy Type']
train=pd.get_dummies(train,columns=one_hot_cols,drop_first=True,dtype=int)
```

```python
train
```

```python
train.columns
```

```python
train['Smoking Status']=le.fit_transform(train['Smoking Status'])
```

```python
train['Smoking Status']
```

```python
train=pd.get_dummies(train,columns=['Exercise Frequency'],drop_first=True,dtype=int)
```

```python
train
```

```python
train.columns
```

```python
train = pd.get_dummies(train, columns=['Property Type'], drop_first=True, dtype=int)
```

```python
train['Policy Start Date']=pd.to_datetime(train['Policy Start Date'],errors='coerce')
```

```python
train['Policy Year']=train['Policy Start Date'].dt.year
train['Policy Month']=train['Policy Start Date'].dt.month
train=train.drop('Policy Start Date',axis=1)
```

```python
train
```

```python
train.columns
```

```python
print(train['Customer Feedback'].isnull().sum())
print(train['Customer Feedback'].value_counts())
```

```python
train['Customer Feedback']=train['Customer Feedback'].fillna('Unknown')
```

```python
train['Customer Feedback'].isnull().sum()
```

```python
feedback_encoded = pd.get_dummies(train['Customer Feedback'], prefix='Feedback', drop_first=False, dtype=int)
train = pd.concat([train.drop(columns=['Customer Feedback']), feedback_encoded], axis=1)
```

```python
train.shape
```

```python
X=train.drop(['id','Premium Amount'],axis=1)
y=train['Premium Amount']
```

```python
print(X.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=2)
```

```python
print(X_train.columns)
```

```python
print(X_test.shape)
```

```python
print(X_train.shape)
```

```python
print(y_train.shape)
```

```python
print(y_test.shape)
```

```python
y_train.value_counts()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
lin_reg=LinearRegression()
lin_reg.fit(X_train,y_train)
```

```python
y_pred=lin_reg.predict(X_test)
```

```python
y_pred[:5]
```

```python
mae=mean_absolute_error(y_test,y_pred)
mse=mean_squared_error(y_test,y_pred)
rmse=np.sqrt(mse)
r2=r2_score(y_test,y_pred)
print(mae)
print(mse)
print(rmse)
print(r2)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
reg_tree=DecisionTreeRegressor(random_state=2)
reg_tree.fit(X_train,y_train)
y_pred_reg_tree=reg_tree.predict(X_test)
```

```python
print(y_pred_reg_tree)
```

```python
# from sklearn import tree
# import matplotlib.pyplot as plt

# plt.figure(figsize=(20,10))
# tree.plot_tree(reg_tree, feature_names=X.columns, filled=True, fontsize=8,max_depth=2)
# plt.show()
```

```python
mae_reg_tree=mean_absolute_error(y_test,y_pred_reg_tree)
mse_reg_tree=mean_squared_error(y_test,y_pred_reg_tree)
rmse_reg_tree=np.sqrt(mse_reg_tree)
r2_reg_tree=r2_score(y_test,y_pred_reg_tree)
print(mae_reg_tree)
print(mse_reg_tree)
print(rmse_reg_tree)
print(r2_reg_tree)
```

```python
importances=reg_tree.feature_importances_
feat_imp=pd.DataFrame({
    'Feature':X_train.columns,
    'Importance':importances
})
print(feat_imp.head(20))
```

```python
print(importances)
```

```python
from sklearn.ensemble import RandomForestRegressor 
```

```python
rf_reg=RandomForestRegressor(n_estimators=100,random_state=2,n_jobs=-1)
rf_reg.fit(X_train,y_train)
```

```python
y_pred_rf=rf_reg.predict(X_test)
```

```python
mae_rf_reg=mean_absolute_error(y_test,y_pred_rf)
mse_rf_reg=mean_squared_error(y_test,y_pred_rf)
rmse_rf_reg=np.sqrt(mse_rf_reg)
r2_score_rf=r2_score(y_test,y_pred_rf)
print(mae_rf_reg)
print(mse_rf_reg)
print(rmse_rf_reg)
print(r2_score_rf)
```
